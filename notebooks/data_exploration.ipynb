# Data exploration Rossman data

## Set up

```python
# import relevant packages
import os
import pandas as pd
import numpy as np
```

```python
# set up paths
data_dir = "../data"
```

## Import and merge data

```python
train_df = pd.read_csv(os.path.join(data_dir, 'train.csv'))
store_df = pd.read_csv(os.path.join(data_dir, 'store.csv'))
```

```python
print(train_df.columns)
print(train_df.shape)
train_df
```

```python
print(store_df.columns)
print(store_df.shape)
store_df
```

```python
# Merge Store data with Train data
data = train_df.merge(store_df, how="left", left_on='Store', right_on='Store')
print(data.columns)
print(data.shape)
```

## Data cleaning

```python
# initial missing values
data.isnull().sum()
```

### Drop rows with missing or zero sales

```python
# check how many rows with missing or zero sales
print("There are {} rows with missing sales data" .format(data.loc[:, "Sales"].isnull().sum()))
print("There are {} rows with zero sales" .format((data["Sales"] == 0).sum()))
```

```python
# drop all rows with missing and zero sales
df_dropsale = data.dropna(subset=["Sales"])
df_dropsale = df_dropsale.loc[~(df_dropsale["Sales"] ==0)]
df_dropsale.shape
```

```python
# Missing values after dropping missing + zero sales
df_dropsale.isnull().sum()
```

### Check datatypes + change

```python
df_dropsale.dtypes
```

```python
df_dropsale.head()
```

```python
# Get relevant date features
df_dropsale["Date"] = pd.to_datetime(df_dropsale["Date"])
df_dropsale["Month"] = df_dropsale["Date"].dt.month
df_dropsale["Year"] = df_dropsale["Date"].dt.year
df_dropsale["Weekday"] = df_dropsale["Date"].dt.dayofweek 
```
